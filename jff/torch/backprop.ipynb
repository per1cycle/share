```python
# Gradient 
def f(x):
    return x ** 2 + x + 1
```

```python
%matplotlib inline
import numpy as np
x = np.arange(-10, 10, 0.1)
y = f(x)
```

```python
import matplotlib.pyplot as plt
plt.plot(x, y)
```

```python
delta = 0.00001
val = 3
print((f(val + delta) - f(val)) / delta)
```

```python
class Val:
    def __init__(self, data, child=()):
        self.data = data
        self.grad = None
        self.prev = set(child)

    def __add__(self, other):
        return Val(self.data + other.data, (self, other))    
    def __mul__(self, other):
        return Val(self.data * other.data, (self, other))
    def __pow__(self, other):
        return Val(self.data ** other.data)
    
    def __repr__(self):
        return f"Value: {self.data}"
    
```

```python
a = Val(1)
b = Val(2)
c = a + b
print(c.prev)
```
